```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

# Metrics

```python
df = pd.read_csv("../output/metrics.csv")
```

## Variando k para t = 100

```python
def plot(t, metric):
    images = ['n01','n02','s01','s02','t01','t02','p01']
    labels = ['N1','N2','S2','S2','T1','T2','P1']
    colors = ['#3867d6', '#4b7bec', '#20bf6b', '#26de81', '#eb3b5a', '#fc5c65', '#8854d0']

    data = [ df[(df.metric == metric) & (df.cover_image_name == image) & (df.t == t)] \
      for image in images ]

    for i in range(0, len(labels)):
      plt.plot('k', 'value', data=data[i], label=labels[i], color=colors[i])
    plt.legend()
    plt.savefig("varying-k-{}-t{}.png".format(metric,t), dpi=200)
```

### MSE

```python
plot(100, 'mse')
```

### PSNR

```python
plot(100, 'psnr')
```

### SSIM

```python
plot(100, 'ssim')
```
